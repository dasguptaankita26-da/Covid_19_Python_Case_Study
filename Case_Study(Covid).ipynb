Q5.2: Compare the recovery rates (recoveries/confirmed cases) between Canada and Australia as of December 31, 2020. Which country showed better management of the pandemic according to this metric?

```python
a=confirm.loc[confirm['Country/Region'].isin(['Canada', 'Australia']),['Country/Region','12/31/20']].groupby('Country/Region').sum()
a.index.name=None
a.columns=['Total Confrim Cases']
a
```

```python
b=recover.loc[recover['Country/Region'].isin(['Canada','Australia']),['Country/Region','12/31/20']].groupby('Country/Region').sum(0)
b.index.name=None
b.columns=['Total Recoveries']
b
```

```python
c=(b['Total Recoveries']/a['Total Confrim Cases'])*100
c.columns=['recovery rate']
c
plt.bar(c.index,c[0])
```

Q5.3: What is the distribution of death rates (deaths/confirmed cases) among provinces in Canada? Identify the province with the highest and lowest death rate as of the latest data point.

```python
a=confirm.loc[confirm['Country/Region'].isin(['Canada']),['Province/State','5/29/21']]
a.set_index('Province/State',inplace=True)
a.columns=['Confirm Cases']
a

```

```python
b=death.loc[confirm['Country/Region'].isin(['Canada']),['Province/State','5/29/21']]
b.set_index('Province/State',inplace=True)
b.columns=['Death Cases']
b
```

```python
import numpy as np
death_rate=(b['Death Cases']/a['Confirm Cases'])
death_rate=death_rate.replace((np.inf,-np.inf),0)
death_rate.plot(kind='bar')
```
